# Tutorial for Reading structures from Protein Data Bank (PDB) files and Generating Inputs for NERDSS

## 1. Introduction

This function enables users to read structures from Protein Data Bank (PDB) file and extract the coordinate information stored inside, and generate NERDSS-readable input files (.mol and .inp files). The functions introduced in this tutorial is a series of step-by-step seperate functions that can be used in Jupyter Notebook. This tutorial shows how to run the Real PDB functions as static functions. It also supports creating "ProteinComplex" objects and run the same funcitons as methods of the objects. For tutorials on that, please refer to the "ioNERDSSUserGuide" in the same repository. For the input requirements of all the functions below, please also refer to the 'ioNERDSSUserGuide'.

## 2. Functions

Import package 'ioNERDSS' that contains all analytical functions.

```python
import ioNERDSS as io
```

### 2.1 Reading PDB File

This function will extract the coordinate information stored inside a PDB file and calculate the COM of each unique chain, as well as recognize the binding information between each pair of chains (all atoms of different unique chains that are closer that 3.0 angstroms are considered as binded), including whether two chains are binded and the coordinates of each binding interface. All the information will be printed on the screen and the returns will contain all the information for further analysis.

```python
Pt_1utc = io.dtb_PDB_separate_read(FileName='1utc.pdb')
```

### 2.2 Filtration of Chains (Optional)

This function will filter the desired chain according to the input list of chain and exclude all the unnecessary coordinate information for future analysis. Here, only chain A, B, and P are being remained so that the number of remaining interaction site will be three. The code below will only show the way to call and use this function, however, the filtered result will not be used for future analysis. 

```python
Pt_1utc_filtered = io.dtb_PDB_filter(Result=Pt_1utc, ChainList = ['A', 'B', 'P'])
```

### 2.3 Visualizing Model (Optional) 

#### 2.3.1 Obtaining IONERDSS generated files in .pdb format

This function will generate an output file in .pdb format that only contains the calculated COMs and reaction interfaces for visualization and comparison with the original PDB file. The input will be the returns of the previous function.

Note: The unit for the coordinates in files in .pdb format is in Angstrom but not nm, so the value will be 10 times larger than that in NERDSS input files.

```python
io.dtb_PDB_write_PDB(Result=Pt_1utc_filtered)
```

#### 2.3.2 Generating 3D Plot 

This function will generate a 3D plot indicaiting the spacial geometry of each simplified chain. The solid lines of different colors are connecting the COM with interfaces within each chain; the black dotted line are connecting each pair of interfaces and the COMs are shown as solid points with their names above. To interact with the plot, other IDEs rather than Jupyter Notebook (such as VSCode) are recommended.

```python
io.dtb_PDB_3D_plot(Result=Pt_1utc)
```

### 2.4 Changing Sigma 

This function allows users to change the value of sigma (the distance between two binding interfaces). The new sigma value and the corresponding coordinates of interfaces will be shown on the screen and the returns will contain all the information for further analysis.

There are 5 pairs of interfaces in this protein and the sigma value is desired to be changed as: A & Q: 0.8 nm and the rest is 0.5 nm, so that the input can be assigned as below.

```python
Pt2_1utc = io.dtb_PDB_change_sigma(Result=Pt_1utc, ChangeSigma=True, SiteList=[0, 3], NewSigma=[0.5, 0.8])
```

Model visualization after changing sigma.

```python
io.dtb_PDB_3D_plot(Result=Pt2_1utc)
```

### 2.5 Calculate Associate Angle

This function will calculate the 5 associating angles of each pair of interfaces. The default normal vector will be assigned as (0, 0, 1). If the co-linear issue occurs, the system will use (0, 1, 0) instead to resolve co-linear issue. The calculated 5 angles will be shown on the screen automatically. If user intends to manually input the normal vector, please refer to function real_PDB_UI, the separated function does not support manual inputs. The returns will contain all the information for further analysis.

```python
result1 = io.dtb_PDB_calc_angle(Result=Pt_1utc)
```

### 2.6 Normalizing COM

This function will normalize the COM of each chain as (0, 0, 0). The interface of each chain will be subtracted by the COM coordinates accordingly. Once the calculation is completed, there will be message shown on the screen. The returns will contain all the information for further analysis.

```python
result2 = io.dtb_PDB_norm_COM(Result=result1)
```

### 2.7 Writing Files

This function will write .inp and .mol files according to all the calculations and modifications above. Multiple .mol file and a .inp file can be found in the same directory as the Jupyter Notebook file once the function finish running.

```python
io.dtb_PDB_write_input(Result=Pt_1utc)
```

## 3. Simulation

To run the simulation, copy the 'nerdss.exe' file generated in the first step in the same directory as .inp and .mol file. Open the same directory in terminal and run command below in system command line to start simulation:

> ./nerdss -f parm.inp
